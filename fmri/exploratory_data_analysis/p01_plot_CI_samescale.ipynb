```python
# Author: Davide Aloi
# Description: plot CI results (coords -25, -33, 37) for patient 01, sham and anodal.
# CI values are obtained from SPM by opening the contrast move > relax for each joystick session.
# Then you need to click on plot --> Confidence interval. 
# Then you select the coordinates and finally you need to access the variable contrast
# and note down the value in field contrast and that in field interval, and insert it in
# data and intervals.

data = [-2.16, -1.33, -0.52, -1.80, # sham day 1 pre/post, day 5 pre/post (in this order)
         1.51, -0.70, -1.23, 2.28, # anodal (...)
         0.2108, 0.3809, -1.046, -3.055] # cathodal

intervals = [3.67, 2.46, 2.54, 3.18, # sham day 1 pre/post, day 5 pre/post (in this order)
             6.51, 5.76, 3.16, 4.36, # anodal
             2.6657, 2.2716, 1.5760, 2.803] # cathodal

conditions = ['S Day 1 pre', 'S Day 1 post','S Day 5 pre','S Day 5 post',
              'A Day 1 pre', 'A Day 1 post','A Day 5 pre','A Day 5 post',
              'C Day 1 pre', 'C Day 1 post','C Day 5 pre', 'C Day 5 post']
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

y_pos = np.arange(len(conditions))

fig, ax = plt.subplots()
ax.yaxis.grid(True)

ax.bar(y_pos, data, yerr=intervals, align='center', alpha=0.58, ecolor ='black', capsize=10)
plt.xticks(y_pos, conditions)
plt.ylabel('CI')
ax.set_title('Condition')

plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('CI_p01_w02_w04.png', dpi = 300)

plt.show()
```

```python
plt.scatter(np.arange(0,12,1), data)  
plt.errorbar(np.arange(0,12,1), data, yerr=intervals, fmt="o")

# Joining dots
n = 0
for i in range(0,6):
    plt.plot([n,n+1], [data[n],data[n+1]],  linestyle="--", color = 'r')
    n += 2

plt.xticks(y_pos, conditions)
plt.ylabel('CI')
plt.title('Condition')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
