```python
# Author Davide Aloi
# Description: plot dcm results patient 01
import numpy as np
import glob
from nilearn import image
import matplotlib.pyplot as plt
import scipy.io

path_DCM = 'C:\\Users\\davide\\Documents\\GitHub\\raindrop_analyses_fmri_eeg\\fmri\\DCM\\dcm_results_p01\\GCM_fullABC.mat'
DCM_results = scipy.io.loadmat(path_DCM)

## Parameters
# posterior probability threshold used
PP_threshold = 0.5

# function to convert diagnoal values to Hz and recenter on 0
funcHz = lambda x: -0.5*(np.exp(x))+0.5 if x != 0 else 0

# Matrix plot xticklabels / yticklabels
labelsUp = ['M1 ↓', 'TH ↓', 'SMA ↓']
labelsLeft = ['M1 ← ', 'TH ← ', 'SMA ←']

# Sessions names
sessions = ['Sham - Day-1 pre stimulation',
            'Sham - Day-1 post stimulation',
            'Sham - Day-5 pre stimulation',
            'Sham - Day-5 post stimulation',
            'Anodal - Day-1 pre stimulation',
            'Anodal - Day-1 post stimulation',
            'Anodal - Day-5 pre stimulation',
            'Anodal - Day-5 post stimulation',
            'Cathodal - Day-1 pre stimulation',
            'Cathodal - Day-1 post stimulation',
            'Cathodal - Day-5 pre stimulation',
            'Cathodal - Day-5 post stimulation']
```

```python
# i = session (1 = w02 day 1 task baseline, 2 = w02 day 1 task post etc)
for i in range(0, 12):
    print(sessions[i])
    title = sessions[i]
    # EP values session i
    this_EP_values = DCM_results['GCM'][i][0]['Ep'][0][0][0][0][0]
    # PP values session i    
    this_PP_values = DCM_results['GCM'][i][0]['Pp'][0][0][0][0][0]

    
    # Converting EP diagnoal to HZ
    for x in range(0,3):
        this_EP_values[x,x] = funcHz(this_EP_values[x, x])


    # Applying threshold to remove EP vlaues where PP < PP_threshold
    this_EP_values = np.where(this_PP_values > PP_threshold, this_EP_values, 0)
    print('EP values')
    print(this_EP_values)
    print('PP values')
    print(this_PP_values)

    ## PLOT thresholded EP values
    plt.rcParams.update({'font.size': 15})
    plt.rcParams["figure.figsize"] = (8,8)  
    
    from nilearn.plotting.cm import _cmap_d as nilearn_cmaps
    cm =  nilearn_cmaps['bwr'] #or plasma, coolwarm. cold_white_hot, bwr
    plt.matshow(this_EP_values, cmap = cm)
    plt.title(sessions[i] + '\n')
    c = plt.colorbar()
    plt.clim(-0.08, 0.08) 
    for (i, j), z in np.ndenumerate(this_EP_values):
        if z != 0:
            plt.text(j, i, '{:0.4f}'.format(z), ha='center', va='center')
    plt.xticks([0,1,2], labelsUp)

    plt.yticks([0,1,2], labelsLeft)
    

    plt.savefig(title + '.jpg', dpi = 240)
    plt.show()
    plt.close()
```
